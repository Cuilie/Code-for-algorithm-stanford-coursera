## Week2 code1 Dijkstra's shortest-path algorithm
The algorithm is designed for computing shortest path from one specific vertex to other vertices among a non-negative directed acyclic graph. The BFS also can be used for computing shortest path, but it has an assumption that all length of edges should be one.


More advanced implementation involved using a customized heap data structure, which is not involved in this submission.

```python
class Dijkstra(Model):
    def __init__(self):
        super().__init__()

        
    def preprocess(self):
        # No need preprocessing for this algorithm
        self.data = self.dataLoader.data
        self.codedict = {}
    
        
    def model(self):
        source_vertex = 1

        # Initialize data structure
        number_of_vertices = max([i[0] for i in m.data])  #index start from 1

        # Initialize processed set
        X = set([source_vertex])

        # Initialzie all vertices set
        V = set(range(1,number_of_vertices+1))

        # Initialize shorest distance list, index is vertex, value is shortest distance
        A = [0 for _ in range(0,number_of_vertices+1)]
        
        # Initialize PATH list, which is only for explanation
        B = [str(source_vertex) for _ in range(0,number_of_vertices+1)]

        # Main loop
        while X != V:

            scorelist = []
            # filter out all edge that head is in X but tail is not
            for i in m.data:
                if i[0] in X:
                    for edge in i[1:]:
                        if edge[0] not in X:
                            scorelist.append([i[0],edge[0], A[i[0]] + edge[1]])

            # compute the shortest one
            selected_edge = min(scorelist,key = lambda x:x[2])

            # updating new vertex 
            old_vertex = selected_edge[0]
            nex_vertex = selected_edge[1]

            #updating A,B
            A[nex_vertex] = selected_edge[2]
            B[nex_vertex] = B[old_vertex] + ('->' + str(selected_edge[1]))
            X.add(nex_vertex)

#             print("Adding ", nex_vertex," to the set X, distance:" , selected_edge[2] )
        
        for i in range(len(A[1:])):
            print("Vertex:",A[1:][i],", PATH:", B[1:][i])
        
        # this is the part for the quiz submission
        temp = [A[i]  for i in range(201) if i in [7,37,59,82,99,115,133,165,188,197]] 
        result = ""
        for i in temp:
            result += ',' + str(i)
        return result[1:]
```

```python
processLine = lambda x : [int(i) if i.isdigit() else [int(j) for j in i.split(",")] for i in map(str.strip,x.split())]
arg = {'fileName':'./data/dijkstraData.txt','numLines':None ,  'processLine' : processLine  }

d = DataLoader(**arg)
m = Dijkstra()
s = Setting(d,m,False)
s.run()
```
