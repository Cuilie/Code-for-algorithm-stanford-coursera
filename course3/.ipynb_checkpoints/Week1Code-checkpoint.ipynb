```python
import pandas as pd
import numpy as np
import os
```

```python
with open('job.txt','r') as f:
    data = f.read()
data = data.split('\n')
entries = data.pop(0)
entries = data.pop(-1)
```

```python
# 1st question
data = np.array([list(map(int, i.split())) for i in data])
data_id = dict(zip(range(len(data)),data))
ko_list = list(zip(range(len(data)),  data[:,0] - data[:,1], data[:,0]    ))
sort_ko = sorted(ko_list, key = lambda x:(x[1],x[2]), reverse = True)
order = [i[0] for i in sort_ko]
orderarray = np.array([list(data_id[i]) for i in order])
cum_list= np.array([sum(orderarray[:,1][:idx+1]) for idx,data in enumerate(orderarray)])
sum(cum_list * orderarray[:,0])
```

```python
# 2st question
data = np.array([list(map(int, i.split())) for i in data])
data_id = dict(zip(range(len(data)),data))
ko_list = list(zip(range(len(data)),  data[:,0]/ data[:,1], data[:,0]    ))
sort_ko = sorted(ko_list, key = lambda x:(x[1],x[2]), reverse = True)
order = [i[0] for i in sort_ko]
orderarray = np.array([list(data_id[i]) for i in order])
cum_list= np.array([sum(orderarray[:,1][:idx+1]) for idx,data in enumerate(orderarray)])
sum(cum_list * orderarray[:,0])
```

```python
with open('link.txt','r') as f:
    data = f.read()

data = data.split('\n')
entries = data.pop(0)
data.pop(-1)
data = np.array([list(map(int, i.split())) for i in data])
```

```python
# initize
v_n_set = []
v_n_set.append(1)


# generate id for edges
data_id = list(zip(range(1,len(data)+1),data))
all_v = set(data[:,0]).union(set(data[:,1]))
cost = 0
terminate = False
```

```python
# select edge thats start from u
while terminate == False:
    
    
    selected_edge = data[np.isin(data[:,0], v_n_set)]
    selected_edge = selected_edge[np.invert(np.isin(selected_edge[:,1], v_n_set))]
    if selected_edge[:,2].shape[0] == 0:
        min_cost1 = 1e7
    else:
        min_cost1 = min(selected_edge[:,2])
        
    min_selected_edge1 = selected_edge[selected_edge[:,2] == min_cost1]


    
    selected_edge = data[np.isin(data[:,1], v_n_set)]
    selected_edge = selected_edge[np.invert(np.isin(selected_edge[:,0], v_n_set))]
    if selected_edge[:,2].shape[0] == 0:
        min_cost2 = 1e7
    else:
        min_cost2 = min(selected_edge[:,2])
    min_selected_edge2 = selected_edge[selected_edge[:,2] == min_cost2]

    
    
    if min_cost1 < min_cost2:
        min_selected_edge = min_selected_edge1
        v_n_set.append(min_selected_edge[0][1])
    else:
        min_selected_edge = min_selected_edge2
        v_n_set.append(min_selected_edge[0][0])
        
    cost = cost + min_selected_edge[0][2]
    new_v = set(v_n_set)
    terminate = (all_v - new_v == set())
    print(cost)
#     break
    


# delete_index = np.where(data == min_selected_edge)
```
